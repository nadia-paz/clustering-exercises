```python
def get_zillow():
    query = '''
    select * from  properties_2017
    join predictions_2017 using(parcelid)
    where transactiondate < '2018' and propertylandusetypeid = 261;
    '''
    df = pd.read_sql(query, get_db_url('zillow'))
    df = df.drop(columns='id')
    return df
```
