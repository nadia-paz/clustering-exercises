```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
from pydataset import data

from sklearn.model_selection import learning_curve
from sklearn.cluster import KMeans

from importlib import reload

pd.options.display.float_format = '{:,.3f}'.format
```

```python
df = data('iris')
```

```python
df
```

```python
# meke columns more readable
new_col = []
for col in df.columns:
    new_col.append(col.lower().replace('.', '_'))
```

```python
new_col
```

```python
# reassign columns
df.columns = new_col
```

```python
df.head(1)
```

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
```

```python
X = df[['petal_length', 'petal_width']]
```

## Modeling

```python
kmeans3 = KMeans(n_clusters=3)
kmeans3.fit(X)

y3 = kmeans3.predict(X)
```

```python
df['cluster_3'] = y3
```

```python
df
```

```python
colors = ['red', 'blue', 'green']
palette = sns.set_palette(sns.color_palette(colors))
plt.figure(figsize=(12, 8))
#plt.subplot(121)
sns.scatterplot(x='petal_length', y='petal_width', data=df, hue='cluster_3', palette=palette, s=300, alpha=0.3)
#plt.subplot(122)
sns.scatterplot(x='petal_length', y='petal_width', data=df, hue='species', hue_order=['setosa', 'virginica', 'versicolor'], palette=palette, s=50)
plt.show()
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(9, 6))
    pd.Series((KMeans(k).fit(X).inertia_ for k in range(2, 12))).plot(marker='D')
```
